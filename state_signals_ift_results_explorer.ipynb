```python
import pickle
import os
import pprint
from ift_results import IFT_Results
```

```python
design_exp_dirs = [
    "SHA512",
    "aes_opencores",
    "aes_secworks",
    "CVA6-DIV",
    "FWRISC-MDS",
    "ZipCPU-DIV",
    "cv32e40s",
    "BOOM"
]
```

```python
for dir_i, selected_design_dir in enumerate(design_exp_dirs):
    print(f"[START] Processing FastPath Experiment #{dir_i:02}: {selected_design_dir}")
    print()

    dir_files = [os.path.join(selected_design_dir, f) for f in os.listdir(selected_design_dir) if os.path.isfile(os.path.join(selected_design_dir, f))]
    ift_files = [f for f in dir_files if "state_and_other_signals_ift_results" in f and ".pkl" in f]

    if len(ift_files)>0:
        for ift_file in ift_files:
            print(f"[INFO ] State Signal IFT Results for File: {ift_file}")
            my_ift_r = None
            with open(ift_file, 'rb') as f:
                my_ift_r_version_unknown = pickle.load(f)
            
                #Convert to latest IFT_Results object (backwards compatible)
                my_ift_r = IFT_Results(my_ift_r_version_unknown.__dhsn_to_shsn__,
                                       my_ift_r_version_unknown.__shsn_to_tvd__ ,
                                       my_ift_r_version_unknown.__ipn_to_isdhsn__,
                )


            pprint.pprint(my_ift_r.get_dhsn_to_influencing_isdhsn(False, False, True))

            print()
    else:
        print("[INFO ] No State Signal IFT Results to Report (HFG Analysis Yielded No Structural Connection)")
        print()

    print(f"[END  ] Processing FastPath Experiment #{dir_i:02}: {selected_design_dir}")

    print("\n\n")
    print("############################################################################")
    print("\n\n")
```
